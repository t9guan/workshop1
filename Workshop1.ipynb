## üîç Core Concept: The NumPy Array

Matplotlib works around two main objects:

- **Axes** ‚Üí The individual chart or subplot inside the figure


# üìä Understanding Regression Evaluation Metrics

When we build a regression model (like predicting **Exam Scores**),  
we need a way to **measure how accurate** our predictions are.  
Here are the most common **statistical error metrics** used to evaluate regression models:

---

## üßÆ 1. Mean Squared Error (MSE)

**Formula:**
$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

**Meaning:**  
- It calculates the **average of the squared differences** between actual values (`y`) and predicted values (`≈∑`).  
- The **larger the MSE**, the worse the model‚Äôs predictions.  
- Squaring emphasizes **larger errors** ‚Äî one big mistake counts a lot.

**Example:**  
If predicted exam scores differ greatly from actual scores, MSE becomes high.

---

## üßæ 2. Root Mean Squared Error (RMSE)

**Formula:**
$$
RMSE = \sqrt{MSE}
$$

**Meaning:**  
- RMSE is simply the **square root of MSE**.  
- It puts the error **back into the same units** as the predicted variable (e.g., exam points).  
- Easier to interpret ‚Äî if RMSE = 3.5, on average, predictions are off by about 3.5 points.

---

## üìâ 3. Mean Absolute Error (MAE)

**Formula:**
$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

**Meaning:**  
- Calculates the **average absolute difference** between predictions and actual values.  
- Treats all errors equally (no squaring).  
- More **robust to outliers** than MSE or RMSE.

---

## üìà 4. R-squared (Coefficient of Determination)

**Formula:**
$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

where  
- $ SS_{res} = \sum (y_i - \hat{y}_i)^2  ‚Üí residual sum of squares  $
- $ SS_{tot} = \sum (y_i - \bar{y})^2  ‚Üí total sum of squares  $

**Meaning:**  
- Measures **how much of the variation** in the dependent variable can be explained by the model.  
- Value ranges from **0 to 1**:
  - **1.0** ‚Üí perfect prediction  
  - **0.0** ‚Üí no predictive power  

**Example:**  
An R¬≤ of 0.82 means 82% of the variation in exam scores is explained by the model.

---

## üß† Summary Table

| Metric | Measures | Range | Lower is Better? | Notes |
|--------|-----------|--------|------------------|-------|
| **MSE** | Average squared error | ‚â• 0 | ‚úÖ Yes | Sensitive to large errors |
| **RMSE** | Root of MSE | ‚â• 0 | ‚úÖ Yes | Same units as output |
| **MAE** | Average absolute error | ‚â• 0 | ‚úÖ Yes | More robust to outliers |
| **R¬≤** | Variance explained | 0 ‚Üí 1 | ‚ùå No | Closer to 1 = better fit |
